```python
# import libraries
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
#softmax function is a linear transformation in logspace, the output is a series that sums to 1
```

```python
# transformation using softmax 
z = np.random.randint(-5,high=15,size=25)
num = np.exp(z)
den = np.sum( np.exp(z) )
sigma = num / den

print(sigma)
print(np.sum(sigma))
```

```python
# plot result
plt.plot(z,sigma,'go')
plt.xlabel('Original number (z)')
plt.ylabel('Softmax Transformed $\sigma$')
plt.yscale('log')
plt.title('$\sum\sigma$ = %g' %np.sum(sigma))
plt.show()
```

```python
#softmax using torch.nn

# create an instance of the class
softer = nn.Softmax(dim=0)

# then apply the data to that function
sigmaT = softer( torch.Tensor(z) )

# now we get the results
print(sigmaT)
```

```python
#def a function for factoring 4!
def factorial(input):
    factored = 1 
    while input > 1:
        factored *= input
        input = input -1
    return factored 
```

```python
factorial(4)
```

```python
#class model 
class model1(object):

  # constructor method
  def __init__(self,numlayers,numunits,name):
    self.layers  = numlayers
    self.units   = numunits # these are attributes
    self.name    = name
    self.weights = np.random.randn(numlayers,self.units)
  
  # other methods
  def sumUnits(self):
    totalUnits = self.layers * self.units
    print(f'There are {totalUnits} units in the model.')

  def trainTheModel(self,x,y):
    self.weights = self.weights*x + y

  def __str__(self):
    return f'This is a {self.name} type of architecture.'
```

```python
instance1 = model1(5,8,'conv neural network')
print(instance1.weights)
```

```python
#cross entropy: the relationship btw two probability distributions 
# all probs must sum to 1!
p = [   1,0   ] # 'true' distribution
q = [ .25,.75 ] # predicted distribution 

H = 0
for i in range(len(p)):
  H -= p[i]*np.log(q[i]) #cross entropy function 

print('Cross entropy: ' + str(H))
```
